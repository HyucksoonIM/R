```r
rowMeans(temp_Test, na.rm = T)
```

```r
best <- Test %>%
    select(id, MEAN) %>%
    arrange(desc(MEAN)) %>%
    head(3)
best
```

```r
ggplot(best, aes(x = reorder(id, -MEAN), y = MEAN, fill=id)) +
    geom_col(width = 0.7) +
    ylim(0,100)
```
